```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import datetime
```

```python
df = pd.read_csv('train.csv')

df.head()
```

```python
tsids = df['tsID'].unique()
tsids
```

```python
df['y'] = df['ACTUAL']
df['ds'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')
df.head()
```

```python
forecast = pd.DataFrame()
for i in tsids:
    m = Prophet(daily_seasonality=True)
    data = df.loc[df['tsID'] == i]
    not_nulls = (data['ACTUAL'].notnull())
    m.fit(data[not_nulls])
    forecast = forecast.append(m.predict(data[~not_nulls]))
```

```python
ids = []
for i in tsids:
    ids += df.loc[df['tsID']==i, 'ID'][-30:].tolist()
forecast['ID'] = ids
```

```python
forecast.head()
```

```python
result = pd.read_csv('SubmissionSample.csv')
result.head()
```

```python
result = pd.merge(result, forecast[['yhat', 'ID']], how='left', on='ID')
```

```python
result.head()
```

```python
result['PREDICTED'] = result['yhat']
result.head()
```

```python
# save forecast to disk for further submission
result[['ID', 'PREDICTED']].to_csv('result.csv', index=False)
```
